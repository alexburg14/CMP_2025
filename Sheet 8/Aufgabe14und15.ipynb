```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data from diamond.freq.gp
data = np.loadtxt('diamond.freq.gp')

# First column is the q-point coordinate
q_points = data[:, 0]

# Remaining columns are the phonon frequencies
frequencies = data[:, 1:]
```

## Aufgabe 1.5

```python
# Plot the phonon dispersion
plt.figure(figsize=(10, 6))

# Plot each phonon branch
for i in range(frequencies.shape[1]):
    plt.plot(q_points, frequencies[:, i], linewidth=2)

plt.xlabel('q-point coordinate', fontsize=12)
plt.ylabel('Frequency (cm$^{-1}$)', fontsize=12)
plt.title('Phonon Dispersion of Diamond', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
