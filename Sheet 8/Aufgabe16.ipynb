```python
import numpy as np
import matplotlib.pyplot as plt

# Load total DOS and PDOS from matdyn.dos
dos_data = np.loadtxt('matdyn.dos')
freq_cm = dos_data[:, 0]
dos_total = dos_data[:, 1]
pdos_atom1 = dos_data[:, 2]
pdos_atom2 = dos_data[:, 3]

# Plot DOS
plt.figure(figsize=(15, 10))
plt.plot(freq_cm, dos_total, label='Total DOS', color='k')
plt.plot(freq_cm, pdos_atom1, label='PDOS atom 1', color='tab:blue', alpha=0.8)
plt.plot(freq_cm, pdos_atom2, label='PDOS atom 2', color='tab:orange', alpha=0.8)
plt.xlabel('Frequency (cm$^{-1}$)')
plt.ylabel('DOS (states/cm$^{-1}$)')
plt.title('Phonon DOS of Diamond')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
