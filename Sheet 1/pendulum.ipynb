```python
import matplotlib.pyplot as plt
```

```python
def leapfrog(x_i: float, v_minushalf: float, dt: float, func) -> tuple:
    v_half = v_minushalf + dt * func(x_i)
    x_new = x_i + dt * v_half   
    return x_new, v_half

def harmonic_func(x: float) -> float:
    k_over_m = 1.0
    return -k_over_m * x

dt = 0.01
t_max = 10.0
x = 0.1
v_half = 0.3
t = 0.0

x_list = [x]
t_list = [t]
v_list = [v_half]
output_filename = "trajectory.txt"

with open(output_filename, 'w') as f:
    f.write("# Time     Position   Velocity(half-step)\n")
    f.write(f"{t:.2f}   {x:.2f}   {v_half:.2f}\n")
    while t < t_max:
        x, v_half = leapfrog(x, v_half, dt, harmonic_func)
        t = t + dt

        x_list.append(x)
        t_list.append(t)
        v_list.append(v_half)

        f.write(f"{t:.2f}   {x:.2f}   {v_half:.2f}\n")
```

```python
plt.plot(t_list, x_list)
plt.xlabel('Time')
plt.ylabel('Position')
```

```python
plt.plot(t_list, x_list, label = "Position")
plt.plot(t_list,v_list, label = "Velocity")
plt.xlabel('Time')
plt.legend()
```
