```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
```

```python
# load data
data = np.loadtxt('/Users/alexgreck/Documents/Studium M.Sc./VM_Share_Folder/CMP/Homework6/bands.dat.gnu')

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))

fig=plt.figure(figsize=(8,5))

for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))

'''
From Post-processing output file:
     high-symmetry point:  0.0000 0.0000 0.0000   x coordinate   0.0000
     high-symmetry point:  0.5000 0.0000 0.5000   x coordinate   0.7071
     high-symmetry point:  0.3750 0.3750 0.7500   x coordinate   1.1748
     high-symmetry point:  0.0000 0.0000 0.0000   x coordinate   2.0934
'''

# Fermi energy (Here: HOMO level from dimon.out)
plt.axhline(13.3609, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
plt.axhline(17.4518, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)

# High symmetry k-points (check bands_pp.out)
plt.axvline(0, linewidth=0.75, color='k', alpha=0.75)
plt.axvline(0.7071, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.1748, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(2.0934, linewidth=0.75, color='k', alpha=0.5)
# text labels
plt.xticks(ticks= [0, 0.7071, 1.1748, 2.0934], \
           labels=['$\Gamma$', 'X', 'K', '$\Gamma$'], size=12)

plt.ylabel("Energy (eV)", size=12)
plt.text(1.4, 13.7, 'HOMO level', size=10, alpha=0.75)
plt.text(1.4, 17.7, 'Fermi energy', size=10, alpha=0.75)
plt.tight_layout()
plt.show()

#plt.savefig('Bandstructure_Diamond.pdf', dpi=300)
```
