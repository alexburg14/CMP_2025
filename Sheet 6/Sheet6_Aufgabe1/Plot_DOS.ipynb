```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
```

```python
# load data
energy, dos, idos = np.loadtxt('/Users/alexgreck/Documents/Studium M.Sc./VM_Share_Folder/CMP/Homework6/diamond.dos', unpack=True)

# make plot
plt.figure(figsize = (12, 6))
plt.plot(energy, dos, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)', size=12)
plt.ylabel('DOS', size=12)
plt.axvline(x=17.4518, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.axvline(x=13.3609, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-10,25)
plt.ylim(0, )
plt.fill_between(energy, 0, dos, where=(energy < 17.4518), facecolor='red', alpha=0.25)
plt.text(12.5, 1, 'HOMO level', size=10, rotation=90, alpha=0.75)
plt.text(16.5, 1, 'Fermi energy', size=10, rotation=90, alpha=0.75)
plt.tight_layout()
plt.show()

#plt.savefig('DOS_Diamond.pdf', dpi=300)
```
