```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Distance.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(8,5))

plt.vlines(df['Distance'][np.argmin(df['Etotal'])],-3,5, ls='--', color='red', alpha=0.5)

plt.plot(df['Distance'], df['Etotal'], marker='o', markersize=3)
plt.xlabel('Atomic distance [A]', size=12)
plt.ylabel('Total energy [Ry]', size=12)
plt.grid(ls='--', alpha=0.5)
plt.xlim(0,2)
plt.ylim(-3,5)
```

```python
print('Minimal Energy',np.min(df['Etotal']), 'Ry')
print('Argument Minimal Energy', np.argmin(df['Etotal']))
print('Distance of Min Energy', df['Distance'][np.argmin(df['Etotal'])], 'A')
```
